# Задание 2
С помощью функции linspace из библиотеки Numpy **создайте массив** t из 51 числа от 0 до 10 включительно.
**Создайте массив** Numpy под названием f, содержащий косинусы элементов массива t.
**Постройте линейную диаграмму**, используя массив t для координат по горизонтали,а массив f - для координат по вертикали. Линия графика должна быть **зеленого цвета.**
Выведите **название диаграммы** - 'График f(t)'. Также добавьте **названия для горизонтальной оси** - 'Значения t' и для вертикальной - 'Значения f'.
**Ограничьте график **по оси x значениями 0.5 и 9.5, а по оси y - значениями -2.5 и 2.5.

```python
#массивы
t = np.linspace(0, 10, 51)
f = np.cos(t)
```

```python
#График
plt.plot(t, f, color = "green")
#параметры графика
plt.axis([0.5, 9.5, -2.5, 2.5])
plt.title("График f(t)", fontdict=title_dict)
plt.xlabel("Значения t", fontdict=label_dict)
plt.ylabel("Значения f", fontdict=label_dict)
plt.show()
```

С помощью функции linspace библиотеки Numpy создайте массив x из 51 числа от -3 до 3 включительно.
Создайте массивы y1, y2, y3, y4 по следующим формулам:

y1 = x**2

y2 = 2 * x + 0.5

y3 = -3 * x - 1.5

y4 = sin(x)


```python
#создаём массивы
x = np.linspace(-3,3,51)
y1 = x ** 2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)
```

Используя функцию subplots модуля matplotlib.pyplot, создайте объект matplotlib.figure.Figure с названием fig и массив объектов Axes под названием ax,причем так, чтобы у вас было 4 отдельных графика в сетке, состоящей из двух строк и двух столбцов. 

В каждом графике массив x используется для координат по горизонтали.В левом верхнем графике для координат по вертикали используйте y1,в правом верхнем - y2, в левом нижнем - y3, в правом нижнем - y4.

Дайте название графикам: 'График y1', 'График y2' и т.д.

Для графика в левом верхнем углу установите границы по оси x от -5 до 5.

Установите размеры фигуры 8 дюймов по горизонтали и 6 дюймов по вертикали.

Вертикальные и горизонтальные зазоры между графиками должны составлять 0.3.

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()

fig.set_size_inches(8, 6)
fig.subplots_adjust(wspace=0.3, hspace=0.3)

ax1.set_title('График y1')
ax1.set_xlim([-5, 5])
ax1.plot(x, y1)

ax2.set_title('График y2')
ax2.plot(x, y2)

ax3.set_title('График y3')
ax3.plot(x, y3)

ax4.set_title('График y4')
ax4.plot(x, y4)
plt.show()
```

Работа с датасетом: Credit Card Fraud Detection

Информация:

Датасет содержит данные о транзакциях, совершенных в 09.2013. 492 мошенничеств из 284 807 транзакций. Набор данных сильно несбалансирован, на положительный класс (мошенничество) приходится 0,172% всех транзакций.
Переменные V1-V28, их описания скрыты по соображениямм конфиденциальности.
Не скрытые переменные: "Time", "Amount", "Class"
"Class" переменная ответа: 1 - мошенничество, 0 - нет мошенничества.

Импортируйте библиотеку Pandas, а также используйте для графиков стиль “fivethirtyeight”.

```python
plt.style.use('fivethirtyeight')
creditcard = pd.read_csv('creditcard.csv')
```

Посчитайте с помощью метода value_counts количество наблюдений для каждого значения целевой переменной Class

```python
class_list = creditcard['Class'].value_counts()
print(class_list)
```

Примените к полученным данным метод plot, чтобы построить столбчатую диаграмму

```python
class_list.plot(kind='barh')
plt.show()
```

Постройте такую же диаграмму, используя логарифмический масштаб.

```python
class_list.plot(kind='barh', logx=True)
plt.show()
```

На следующем графике постройте две гистограммы по значениям признака V1 - одну для мошеннических транзакций (Class равен 1) и другую - для обычных (Class равен 0). 

```python
class0 = creditcard.loc[creditcard['Class'] == 0, ['V1']]
class1 = creditcard.loc[creditcard['Class'] == 1, ['V1']]
```

Подберите значение аргумента density так, чтобы по вертикали графика было расположено не число наблюдений, а плотность распределения. Число бинов должно равняться 20 для обеих гистограмм, а коэффициент alpha сделайте равным 0.5, чтобы гистограммы были полупрозрачными и не загораживали друг друга. Создайте легенду с двумя значениями: “Class 0” и “Class 1”. Гистограмма обычных транзакций должна быть серого цвета, а мошеннических - красного. Горизонтальной оси дайте название “V1”.

```python
plt.hist(class0['V1'], bins=20, density=True, alpha=0.5, label='Class 0', color='grey')
plt.hist(class1['V1'], bins=20, density=True, alpha=0.5, label='Class 1', color='red')
plt.legend()
plt.show()
```
